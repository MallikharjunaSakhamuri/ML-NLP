```python
dataset= np.zeros((len(all_reviews), 2)).tolist()
print(type(dataset))
```

```python
j = 0
for i in (positive_reviews):
    dataset[j][0] = (i.text)
    dataset[j][1] = 0 
    j = j + 1
    
for i in (negative_reviews):
    dataset[j][0] = (i.text)
    dataset[j][1] = 1 
    j = j + 1    

print(dataset[100][0])    

dataset = shuffle(dataset)

print(dataset[100][0]) 
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

corpus = []
for i in range(0, len(all_reviews)):
  review = re.sub('[^a-zA-Z]', ' ', dataset[i][0])
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python

Arraycorpus = np.array(corpus)

#print(Arraycorpus)
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
y= np.zeros((len(all_reviews), 1))

for i in range(len(all_reviews)):                  
    y[i] = dataset[i][1]

#Y = data[:,-1]
```

```python
print(len(dataset))
print(len(X))
print(len(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
#classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```
